```python
from transformers import pipeline
notes_generator=pipeline('text-generation',model='gpt2')
chapter=input("enter chapter name:")

prompt_notes = f"""
write detailed study notes for {chapter}
include explanation, example and key points
"""
detailed_notes=notes_generator(
    prompt_notes,
    max_length=300,
    num_return_sequences=1
)
print(detailed_notes[0]['generated_text'])

prompt_summary = f"""
summarize the chapter: {chapter} in one page
"""
summary=notes_generator(
    prompt_summary,
    max_length=200,
    num_return_sequences=1
)
print("\n one-page summary \n")
print(summary[0]['generated_text'])

prompt_definition=f"""
list important definitions and formulas from the chapter:{chapter}
use bullet points
"""
definitions=notes_generator(
    prompt_definition,
    max_length=200,
    num_return_sequences=1
)
print("\n definitions and formulas:")
print(definitions[0]['generated_text'])
```
