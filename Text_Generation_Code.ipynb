```python
from transformers import pipeline
#load text generation model(gen ai)
generator = pipeline(
     "text-generation",
     model="gpt2"
      )
#input prompt
prompt = "artificial intelligence in finanace will"

#generate text
result = generator(prompt, max_length=50, num_return_sequences=5)

print("ai generated text:")
print(result[0]['generated_text'])
```
